```python
from sklearn.datasets import load_breast_cancer
from learn_ml.linear_model.logistics import LogisticsRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
```

## Full-Gradient 梯度下降法

### 没有bias 没有正则化项

```python
clf = LogisticsRegression(optimizer='gd',max_iter=10000)
```

```python
X,y = load_breast_cancer(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred),f1_score(y_test,y_pred),clf.n_iter
```

#### 训练过程可视化

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(clf.n_iter))
```

```python
plt.plot(x,clf.losses)
```

```python
clf.losses
```

### 没有bias 有L2 正则

```python
clf = LogisticsRegression(optimizer='gd',max_iter=1000,penalty='l2',C=1.0)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred),f1_score(y_test,y_pred)
```

#### 训练过程可视化

```python
x = list(range(clf.n_iter))
```

```python
plt.plot(x,clf.losses)
```

## Min-Batch-Gradient 梯度下降

```python
clf = LogisticsRegression(max_iter=1000,lr=0.01,tol=1e-5,batch_size=1)
```

```python
X,y = load_breast_cancer(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred),f1_score(y_test,y_pred),clf.n_iter
```

### 训练过程可视化

```python
x = list(range(clf.n_iter))
```

```python
plt.plot(x,clf.losses)
```
